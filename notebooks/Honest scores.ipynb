# Data load

```python
# datasets
datasets = ("en_binary", "en_queer_nonqueer")
granularities = ("point_k",)
similarity_granularities = ("point_k",)  # similarity is computed point-k only
honest_score_metrics = ("full", "by_identity")
semantic_similarity_metrics = ("by_family", "by_scale", "by_model_scale", "by_family_and_identity", "by_scale_and_identity")

# identities
all_identities = ["girlfriend", "he", "she", "the aunt", "the boy",  "the boyfriend",  "the bride",  "the brother",  "the child",  "the children",  "the daughter",  "the father",  "the female",  "the gentleman",  "the girl",  "the girlfriends",  "the groom",  "the husband",  "the ladies",  "the lady",  "the lords",  "the male",  "the man",  "the men",  "the mother",  "the sister",  "the son",  "the spouses",  "the uncle",  "the wife",  "the wives",  "the woman",  "the women",  "the young"]
female_identities = ["girlfriend", "she", "aunt", "bride", "daughter", "female", "girl", "ladies", "lady", "mother",  "sister", "wife", "wives", "woman", "women"]
male_identities = ["he", "boy", "boyfriend", "brother", "father", "gentleman", "groom", "husband", "male", "man", "men", "son", "uncle"]
neutral_identities = ["child",  "children",  "spouses",  "young"]
young_identities = ["child", "young", "children", "son", "daughter", "boy"]
nonyoung_identities = [i for i in all_identities if i not in young_identities]
queer_identities = ['Fae',  'LGBT person',  'LGBTQ person',  'LGBTQI person',  'LGBTQIA person',  'LGBTQIA+ person',  'They',  'Xe',  'Ze',  'agender',  'androgynous',  'aromantic',  'asexual',  'biromantic',  'bisexual',  'catgender',  'cisgender',  'demigender',  'demisexual',  'enby',  'gay',  'gender conforming',  'gender non-conforming',  'genderfluid',  'genderless',  'genderqueer',  'heteroromantic',  'heterosexual',  'homoromantic',  'homosexual',  'intersexual',  'lesbian',  'non-binary',  'nonqueer',  'pangender',  'panromantic',  'pansexual',  'polygender',  'queer',  'straight',  'trans',  'transgender',  'transman',  'transsexual',  'transwoman',  'xenogender']


# data loading
analysis = dict()
# load honest scores
analysis_file_template = "../src/experiments/honest_{0}.honest_score.{1}.{2}.csv"
analysis["honest_score"] = dict()
for dataset in datasets:
    analysis["honest_score"][dataset] = dict()
    for metric in honest_score_metrics:
        analysis["honest_score"][dataset][metric] = dict()
        for granularity in granularities:
            if metric == "by_identity":
                df = pandas.read_csv(analysis_file_template.format(dataset, metric, granularity))

                # normalize identities
                df.loc[df.identity == "ladies", "identity"] = "lady"
                df.loc[df.identity == "wives", "identity"] = "wife"
                df.loc[df.identity == "women", "identity"] = "woman"
                df.loc[df.identity == "men", "identity"] = "man"
                df_identities = df.identity.values

                # map identities to sex groups
                sex_identities = list()
                for i in df_identities:
                    if i in male_identities:
                        sex_identities.append("male")
                    elif i in female_identities:
                        sex_identities.append("female")
                    else:
                        sex_identities.append("neutral")
                df["sex_identity"] = sex_identities

                # map identities to age groups
                age_identities = list()
                for i in df_identities:
                    if i in young_identities:
                        age_identities.append("young")
                    elif i in nonyoung_identities:
                        age_identities.append("nonyoung")
                    else:
                        age_identities.append("unknown")
                df["age_identity"] = age_identities

                analysis["honest_score"][dataset][metric][granularity] = df
            else:
                analysis["honest_score"][dataset][metric][granularity] = pandas.read_csv(analysis_file_template.format(dataset, metric, granularity))


# load semantic similarity
analysis_file_template = "../src/experiments/honest_{0}.similarity.{1}.{2}.csv"
analysis["semantic_similarity"] = dict()
for dataset in datasets:
    analysis["semantic_similarity"][dataset] = dict()
    for metric in semantic_similarity_metrics:
        analysis["semantic_similarity"][dataset][metric] = dict()
        for granularity in similarity_granularities:
            if metric == "by_identity":
                df = pandas.read_csv(analysis_file_template.format(dataset, metric, granularity))
                
                # normalize identities
                df.loc[df.identity == "ladies", "identity"] = "lady"
                df.loc[df.identity == "wives", "identity"] = "wife"
                df.loc[df.identity == "women", "identity"] = "woman"
                df.loc[df.identity == "men", "identity"] = "man"
                df_identities = df.identity.values

                # map identities to sex groups
                sex_identities = list()
                for i in df_identities:
                    if i in male_identities:
                        sex_identities.append("male")
                    elif i in female_identities:
                        sex_identities.append("female")
                    else:
                        sex_identities.append("neutral")
                df["sex_identity"] = sex_identities

                # map identities to age groups
                age_identities = list()
                for i in df_identities:
                    if i in young_identities:
                        age_identities.append("young")
                    elif i in nonyoung_identities:
                        age_identities.append("nonyoung")
                    else:
                        age_identities.append("unknown")
                df["age_identity"] = age_identities

                analysis["semantic_similarity"][dataset][metric][granularity] = df
            else:
                analysis["semantic_similarity"][dataset][metric][granularity] = pandas.read_csv(analysis_file_template.format(dataset, metric, granularity))
```
